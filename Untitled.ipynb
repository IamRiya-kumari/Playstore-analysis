### Let's define a function to view some more insights of the dataset.

```python
def printinfo():
    temp = pd.DataFrame(index=googleplaystore.columns)
    temp['data_type'] = googleplaystore.dtypes
    temp['null_count'] = googleplaystore.isnull().sum()
    temp['unique_count'] = googleplaystore.nunique()
    return temp
```
